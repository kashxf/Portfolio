#
## **Top Actors:**

#### From the dataset the maximum count is for Anupam Kher, it means that Anupam Kher is the top actor.

```python
"""The Counter class in Python, which is part of the collections module, is a  way to count the occurrences of items
in an iterable (like a list) by storing them as dictionary keys and their counts as dictionary values."""

from collections import Counter
actor_counts = Counter()

for cast_list in data['cast'].dropna():  #iterate throughout the whole cast column
    for actor in cast_list.split(', '): # If there is a comma between the name of the cast theit splits.
        actor_counts[actor] += 1         # incrementing its count by 1

top_actors = actor_counts.most_common(10)  #find the top 10 most common actos (from counter lib)

top_actors_data = pd.DataFrame(top_actors, columns=['Actor', 'Count']) #tranfering this data set into a dataframe conttaining 
                                                              #two columns
top_actors_data.index = top_actors_data.index + 1  # adding 1 so it show the starting from 1 not from default 0 
print(top_actors_data)
```

```python

g = sns.barplot(y='Actor', x='Count', data=top_actors_data, palette="dark:brown", edgecolor="black", width=0.5)

plt.title(" Top Actors (Overall)")
```

#
### **Top Movie Actors**

```python
actor_counts_movies = Counter()

for _, row in data.iterrows():
    if pd.notna(row['cast']):
        actors = row['cast'].split(', ')  # Split actors by comma
        for actor in actors:
            if row['type'] == 'Movie':
                actor_counts_movies[actor] += 1

top_actors_movies = actor_counts_movies.most_common(10)

top_actors_movies_data = pd.DataFrame(top_actors_movies, columns=['Actor', 'Count'])

top_actors_movies_data.index = top_actors_movies_data.index + 1
```

```python
g = sns.barplot(y='Actor', x='Count', data=top_actors_movies_data, palette="dark:green", edgecolor="black", width=0.5)
plt.title("Top Movie Actors")  # Update the title to reflect 'Movie' actors
plt.show()
```

#
### **Top TV Show Actors**

```python
actor_counts_tv = Counter()

for _, row in data.iterrows():
    if pd.notna(row['cast']):
        actors = row['cast'].split(', ')  # Split actors by comma
        for actor in actors:
            if row['type'] == 'TV Show':
                actor_counts_tv[actor] += 1

top_actors_tv = actor_counts_tv.most_common(10)

top_actors_tv_data = pd.DataFrame(top_actors_tv, columns=['Actor', 'Count'])

top_actors_tv_data.index = top_actors_tv_data.index + 1

g = sns.barplot(y='Actor', x='Count', data=top_actors_tv_data, palette="dark:pink", edgecolor="black", width=0.5)
plt.title("Top TV Show Actors")  # Update the title to reflect 'Movie' actors
plt.show()
```
